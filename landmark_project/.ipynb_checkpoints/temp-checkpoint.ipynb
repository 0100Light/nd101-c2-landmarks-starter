```python
import numpy as np
import csv
import pandas as pd

filePath = '03.txt'

tableHead = []
columnCount = 0

with open(filePath) as f:
    r = csv.reader(f, delimiter='\t')
    rowCount = 0
    arr = []
    for row in r:
        if rowCount == 0:
            tableHead = row
            columnCount = len(row)
        else:
            arr.append(row)

        rowCount += 1

    df = pd.DataFrame(arr, columns=tableHead)


df
```

```python
import matplotlib.pyplot as plt

genData = []
fig = plt.figure(figsize=(25,25))

for i in range(columnCount):
    uniqueArray = np.unique(df.iloc[:, i])

    idx = [x for x in range(len(uniqueArray))]
    normalArr = np.random.uniform(0, len(uniqueArray), 200)

    randomIdx = np.floor(normalArr).astype('int')
    ax = plt.subplot(7,10,i+1)
    ax.hist(randomIdx)
    ax.set_title(tableHead[i] + ", " + str(len(uniqueArray)))


    randomData = [uniqueArray[x] for x in randomIdx]
    genData.append(randomData)

plt.show()
```

```python
genData = pd.DataFrame(genData)
genData = genData.transpose()
genData.columns = tableHead
genData
```

## Resolving grouped question problems

```python
def genArrayHist(idxArrayLength, generateCount):
    genIdx = np.random.uniform(0, idxArrayLength, generateCount)
    genIdx = np.floor(genIdx).astype('int')

    fig = plt.figure()
    plt.hist(genIdx)

    return genIdx, fig
```

```python
# Coffee types
idx, plot = genArrayHist(23, 200)
idx, plot
```
